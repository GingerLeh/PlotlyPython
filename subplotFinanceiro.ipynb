```python
import plotly.graph_objects as go
from pandas_datareader import data
from plotly.subplots import make_subplots
```

```python
# Definindo titulos e data
title = ['Close Price', 'Open Price', 'High Price', 'Low Price']
data = data.DataReader(name = 'MGLU3.SA', data_source='yahoo')
```

```python
# Plotando os graficos
from cgitb import text


fig = make_subplots(rows=2, cols=2, subplot_titles=title)

# Close price
fig.add_trace(go.Scatter(
    x = data.index,
    y=data['Close'],
    name= 'Close Price',
    mode='lines',
    line=dict(color = 'blue', width=1)
), row=1, col=1)

# Open price
fig.add_trace(go.Scatter(
    x = data.index,
    y=data['Open'],
    name= 'Open Price',
    mode='lines',
    line=dict(color = 'red', width=1)
), row=1, col=2)

# High price
fig.add_trace(go.Scatter(
    x = data.index,
    y=data['High'],
    name= 'High Price',
    mode='lines',
    line=dict(color = 'gray', width=1)
), row=2, col=1)

# Low price
fig.add_trace(go.Scatter(
    x = data.index,
    y=data['Low'],
    name= 'Low Price',
    mode='lines',
    line=dict(color = 'green', width=1)
), row=2, col=2)

# Para colocar valores monetários
fig.update_yaxes(tickprefix='$')

fig.update_xaxes(showline = True, linewidth=1, linecolor = 'gray')
fig.update_yaxes(showline = True, linewidth=1, linecolor = 'gray')

fig.update_layout(
    title = dict(
        text='MGLU3',
        x = 0.5, 
        y = 0.9),
        font_family = 'Arial', 
        plot_bgcolor = 'white', 
        legend = dict(
            title = 'Options', 
            bgcolor = 'lightgray',
            bordercolor = 'red',
            borderwidth = 1, 
            font = dict(
                size = 10,
                color = 'darkred'
            )
        )
)

fig.show()

```
